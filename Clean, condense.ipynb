```python
import pandas as pd
```

```python
#Import csv, ensure that it is loaded
path = "2019-2023.csv"
df=pd.read_csv(path)
df.head(10)
```

```python
#Verify column names for later cleaning
df.columns
```

```python
#Drop columns that are redundant
df=df.drop(columns=['oty_qtrly_estabs_chg', 'oty_qtrly_estabs_pct_chg',
       'oty_month1_emplvl_chg', 'oty_month1_emplvl_pct_chg',
       'oty_month2_emplvl_chg', 'oty_month2_emplvl_pct_chg',
       'oty_month3_emplvl_chg', 'oty_month3_emplvl_pct_chg',
       'oty_total_qtrly_wages_chg', 'oty_total_qtrly_wages_pct_chg',
       'oty_taxable_qtrly_wages_chg', 'oty_taxable_qtrly_wages_pct_chg',
       'oty_qtrly_contributions_chg', 'oty_qtrly_contributions_pct_chg',
       'oty_avg_wkly_wage_chg', 'oty_avg_wkly_wage_pct_chg'])
df.columns
```

```python
#Combine monthly columns to track quarters
df['total_employed'] = df['month1_emplvl'] + df['month2_emplvl'] + df['month3_emplvl']
df['%chg_total_emp'] = (df['lq_month1_emplvl'] + df['lq_month2_emplvl'] +df['lq_month3_emplvl']) / 3
df=df.drop(columns=['month1_emplvl','month2_emplvl','month3_emplvl','lq_month1_emplvl','lq_month2_emplvl','lq_month3_emplvl'])
```

```python
#Rename columns for ease of reading
new_names={'area_fips' :'area_code','qtr':'quarter','qtrly_estabs':'establishments', 
           'total_qtrly_wages':'total_wages', 'taxable_qtrly_wages':'taxable_wage',
           'avg_wkly_wage':'average_weekly_wage','lq_qtrly_estabs':'%chg_lq_estabishments',
           'lq_total_qtrly_wages':'%chge_lq_total_qtrly_wages','lq_taxable_qtrly_wages':'%chge_lq_taxable_wages',
           'lq_avg_wkly_wage':'%chge_avg_wkly_wage'}
df.rename(columns=new_names, inplace=True)
df.head()
```

```python
#Combine rows by date and area.
grouped_df = df.groupby(['area_code', 'year', 'quarter']).agg({
    'establishments': 'sum',
    'total_wages': 'sum',
    'taxable_wage': 'sum',
    'average_weekly_wage': 'sum',
    '%chg_lq_estabishments': 'sum',
    '%chge_lq_total_qtrly_wages': 'mean',
    '%chge_lq_taxable_wages': 'mean',
    '%chge_avg_wkly_wage': 'mean',
    'total_employed': 'sum',
    '%chg_total_emp':'mean'
}).reset_index()

clean_df = grouped_df[~grouped_df['area_code'].astype(str).str.endswith('000')].copy()
clean_df
```

```python
df=clean_df[['area_code','quarter','year','establishments','total_employed','total_wages','taxable_wage',
    'average_weekly_wage','%chg_lq_estabishments','%chg_total_emp','%chge_lq_total_qtrly_wages',
    '%chge_lq_taxable_wages','%chge_avg_wkly_wage']]
df
```

```python
df.to_csv('CleanData.csv', index=False)
```
