```python
import pandas as pd
```

```python
path='CleanData.csv'
df=pd.read_csv(path)
df.head(10)
```

```python
df["id"] = df.index + 1
df.columns
```

```python
df=df[['id','area_code', 'quarter', 'year', 'establishments', 'total_employed',
       'total_wages', 'taxable_wage', 'average_weekly_wage',
       '%chg_lq_estabishments', '%chg_total_emp', '%chge_lq_total_qtrly_wages',
       '%chge_lq_taxable_wages', '%chge_avg_wkly_wage']]
df.head()
```

```python
#save the df
df.to_csv('CleanData.csv', index=False)
```

```python
#Year
year=df[['id','quarter', 'year']]
year.to_csv('yearDf.csv', index=False)
```

```python
#Industry
industry=df[['id','area_code', 'establishments', 'total_employed',
       'total_wages', 'taxable_wage']]
industry.to_csv('industryDf.csv',index=False)
```

```python
#Wages
wages=df[['id','total_wages', 'average_weekly_wage',
       '%chg_lq_estabishments', '%chg_total_emp', '%chge_lq_total_qtrly_wages',
       '%chge_lq_taxable_wages', '%chge_avg_wkly_wage']]
wages.to_csv('wagesDf.csv', index= False)
```
